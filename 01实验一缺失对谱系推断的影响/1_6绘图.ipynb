# 实验一缺失比例对最终结果的影响绘图

```python
path1_1=r'C:\Users\pro\Desktop\实验一缺失模式对建设的影响研究\1.1缺失比例对结果的影响'
data1=pd.read_csv(os.path.join(path1_1,"normal_50taxa25chara_sim.csv"))
data2=pd.read_csv(os.path.join(path1_1,"normal_50taxa50chara_sim.csv"))
data3=pd.read_csv(os.path.join(path1_1,"normal_50taxa75chara_sim.csv"))
data4=pd.read_csv(os.path.join(path1_1,"normal_50taxa100chara_sim.csv"))
```

```python
indexs=pd.Series([0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45, 0.5,0.55,0.60,0.65,0.70,0.75])
data1.rename(columns={'V1':'NJ', 'V2':'UPGMA', 'V3':'ML','V4':'MP'}, inplace = True)
data1.index = indexs

data2.rename(columns={'V1':'NJ', 'V2':'UPGMA', 'V3':'ML','V4':'MP'}, inplace = True)
data2.index = indexs

data3.rename(columns={'V1':'NJ', 'V2':'UPGMA', 'V3':'ML','V4':'MP'}, inplace = True)
data3.index = indexs

data4.rename(columns={'V1':'NJ', 'V2':'UPGMA', 'V3':'ML','V4':'MP'}, inplace = True)
data4.index = indexs
```

```python
%matplotlib notebook
xticks=[0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75]
#xticks=["10%","20%","30%","40%","50%","60%","70%","80%"]
fig = plt.figure(figsize=(15,10)) #如果不传入参数默认画板1
#fig = plt.figure() #如果不传入参数默认画板1
ax1=plt.subplot(2,2,1)   
data1['NJ'].plot(ax=ax1,linestyle=':',marker='o')
data1['UPGMA'].plot(ax=ax1,linestyle='-',marker='1')
data1['ML'].plot(ax=ax1,linestyle='dashed',marker='s')
data1['MP'].plot(ax=ax1,linestyle='dashed',marker='v')
plt.legend()
ax1.set_title('25chara')
plt.xlabel('缺失率')
plt.ylabel('Similarity')
ax1.set_xticks(xticks)

ax2=plt.subplot(2,2,2)   
data2['NJ'].plot(ax=ax2,linestyle=':',marker='o')
data2['UPGMA'].plot(ax=ax2,linestyle='-',marker='1')
data2['ML'].plot(ax=ax2,linestyle='dashed',marker='s')
data2['MP'].plot(ax=ax2,linestyle='dashed',marker='v')
plt.legend()
ax2.set_title('50chara')
plt.xlabel('缺失率')
plt.ylabel('Similarity')
ax2.set_xticks(xticks)

ax3=plt.subplot(2,2,3)   
data3['NJ'].plot(ax=ax3,linestyle=':',marker='o')
data3['UPGMA'].plot(ax=ax3,linestyle='-',marker='1')
data3['ML'].plot(ax=ax3,linestyle='dashed',marker='s')
data3['MP'].plot(ax=ax3,linestyle='dashed',marker='v')
plt.legend()
ax3.set_title('75chara')
plt.xlabel('缺失率')
plt.ylabel('Similarity')
ax3.set_xticks(xticks)

ax4=plt.subplot(2,2,4)   
data4['NJ'].plot(ax=ax4,linestyle=':',marker='o')
data4['UPGMA'].plot(ax=ax4,linestyle='-',marker='1')
data4['ML'].plot(ax=ax4,linestyle='dashed',marker='s')
data4['MP'].plot(ax=ax4,linestyle='dashed',marker='v')
plt.legend()
ax4.set_title('100chara')
plt.xlabel('缺失率')
plt.ylabel('Similarity')
ax4.set_xticks(xticks)
plt.savefig('图5 数据缺失比例对建树结果的影响.png',bbox_inches='tight',dpi=fig.dpi,pad_inches=0.0)
plt.show()
```
