```python
%matplotlib notebook
fig = plt.figure(figsize=(15,10)) #如果不传入参数默认画板1
#fig = plt.figure() #如果不传入参数默认画板1
ax1=plt.subplot(2,2,1)   
data1['our'].plot(ax=ax1,linestyle=':',marker='o')
data1['Random'].plot(ax=ax1,linestyle='-')
data1['Median'].plot(ax=ax1,linestyle='dashed')
plt.legend()
ax1.set_title('Aguado2009')
plt.xlabel('缺失率')
plt.ylabel('RMSE')

ax2=plt.subplot(2,2,2)
data1['our'].plot(ax=ax2,linestyle=':',marker='o')
data1['Random'].plot(ax=ax2,linestyle='-')
data1['Median'].plot(ax=ax2,linestyle='dashed')
plt.legend()
ax2.set_title('Aria2015')
plt.xlabel('缺失率')
plt.ylabel('RMSE')

ax3=plt.subplot(2,2,3)   
data1['our'].plot(ax=ax3,linestyle=':',marker='o')
data1['Random'].plot(ax=ax3,linestyle='-')
data1['Median'].plot(ax=ax3,linestyle='dashed')
plt.legend()
ax3.set_title('Dikow2009')
plt.xlabel('缺失率')
plt.ylabel('RMSE')

ax4=plt.subplot(2,2,4)
data1['our'].plot(ax=ax4,linestyle=':',marker='o')
data1['Random'].plot(ax=ax4,linestyle='-')
data1['Median'].plot(ax=ax4,linestyle='dashed')
plt.legend()
ax4.set_title('Griswold1999')
plt.xlabel('缺失率')
plt.ylabel('RMSE')
#显示图像
plt.savefig('四个数据集缺失插补RMSE比较.png')
plt.show()
```
